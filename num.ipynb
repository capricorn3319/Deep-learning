```python
import cv2
import time
import numpy as np
import torch

# Load YOLOv5 model
model = torch.hub.load("ultralytics/yolov5", "yolov5s")

# Load classes
classes = []
with open("/home/cmos/Downloads/yolov5/yolov5/data/coco.yaml", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Initialize video stream
cap = cv2.VideoCapture(0)

# Define time periods
time_periods = []
start_time = time.time()
for i in range(10):
    time_periods.append(start_time + i*60)

# Initialize counters
counts = [0]*len(time_periods)
total_counts = 0

while True:
    # Read frame from video stream
    ret, frame = cap.read()
    if not ret:
        break

    # Run object detection on the frame
    outputs = model(frame)

    # Process detections
    boxes = []
    confidences = []
    class_ids = []
    for i in range(len(outputs.xyxy[0])):
        box = outputs.xyxy[0][i].cpu().numpy()
        confidence = outputs.xyxy[0][i, 4].item()
        class_id = int(outputs.xyxy[0][i, 5].item())

        # Filter out weak detections
        if confidence > 0.5 and class_id == 0:
            boxes.append((box[0], box[1], box[2], box[3]))
            confidences.append(confidence)
            class_ids.append(class_id)

    # Apply non-maximum suppression
    indices = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    # Update counters
    if len(indices) > 0:
        indices = indices.flatten()
        for i in indices:
            total_counts += 1
            for j in range(len(time_periods)):
                if time.time() < time_periods[j]:
                    counts[j] += 1
                    break

    # Display counts
    for i in range(len(time_periods)):
        print(f"Count for period {i+1}: {counts[i]}")
    print(f"Total count: {total_counts}")

    # Display output
    cv2.imshow("YOLOv5", frame)
    if cv2.waitKey(1) == ord("q"):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
