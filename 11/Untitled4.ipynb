```python
import hazm
from hazm import word_tokenize, sent_tokenize

# متن ورودی
input_text = "شیر خوردنی لبنیات و شیر حیوان جنگل است.من برای صبحانه شیر میخورم. شیر آب خانه ما خراب است"

# توکن‌سازی ورودی به کلمات و جملات
tokens = word_tokenize(input_text)
sentences = sent_tokenize(input_text)

# استفاده از تابع مخصوص پیش پردازش متون Hazm برای پاک کردن نویزهای متنی
normalizer = hazm.Normalizer()
clean_text = normalizer.normalize(input_text)

# استفاده از واژه‌نامه‌های معنایی برای تعیین معانی کلمات
lemmatizer = hazm.Lemmatizer()
normalized_tokens = []
for token in tokens:
    normal_form = lemmatizer.lemmatize(token)
    normalized_tokens.append(normal_form)

# حذف کلمات بی‌معنی
stopwords_list = hazm.stopwords_list()
meaningful_words = []
for token in normalized_tokens:
    if token not in stopwords_list:
        meaningful_words.append(token)

# تعداد کلمات و جملات
num_words = len(meaningful_words)
num_sentences = len(sentences)

# تعداد تکرار هر کلمه
word_frequency = {}
for word in meaningful_words:
    if word in word_frequency:
        word_frequency[word] += 1
    else:
        word_frequency[word] = 1

# نمایش خروجی پس از پیش پردازش
print("متن ورودی: ", input_text)
print("متن پاک شده: ", clean_text)
print("کلمات: ", tokens)
print("جملات: ", sentences)
print("کلمات نرمال‌شده: ", normalized_tokens)
print("کلمات معنادار: ", meaningful_words)
print("تعداد کلمات: ", num_words)
print("تعداد جملات: ", num_sentences)
print("تعداد تکرار هر کلمه: ", word_frequency)
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set up the data generator for image preprocessing
datagen = ImageDataGenerator(rescale=1./255)

# Set up the training and validation data directories
train_dir = 'path/to/training/data'
val_dir = 'path/to/validation/data'

# Set up the training and validation data generators
train_gen = datagen.flow_from_directory(train_dir, target_size=(224, 224), batch_size=32, class_mode='categorical')
val_gen = datagen.flow_from_directory(val_dir, target_size=(224, 224), batch_size=32, class_mode='categorical')

# Define the model architecture
model = tf.keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model with an appropriate loss function and optimizer
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(train_gen, validation_data=val_gen, epochs=10)

# Save the model for future use
model.save('food_classification_model.h5')
```
